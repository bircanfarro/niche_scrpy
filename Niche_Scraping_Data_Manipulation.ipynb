```python
#When using pd.merge directly to merge the 'grade' and 'badge' datasets the duplicate rows crated a problem 
#the number of rows raised up to 200M, therefore the merge_grade and merge_badge functions were defined.



grade_columns = {
    'Overall Niche Grade': 'Overall Niche Grade',
    'Public Schools': 'Public Schools',
    'Health': 'Health & Fitness',
    'Outdoor Activities': 'Outdoor Activities',
    'Good For Families': 'Good for Families',
    'Safety': 'Crime & Safety',
    'Diversity': 'Diversity',
    'Good For Millennials': 'Good for Millennials',
    'Good For Retirees': 'Good for Retirees',
    'Affordability': 'Cost of Living',
    'Housing Market': 'Housing'
}

def merge_grade(df_from, df_to, key):
    label = grade_columns[key]
    df_some_grades = df_from.loc[df_from.label == label,:]
    df_some_grades = df_some_grades.drop_duplicates()

    df_some_grades = df_some_grades.drop(['guid', 'label', 'rankingGuid'], 1)
    df_some_grades.columns = df_some_grades.columns.str.replace('value', '%s' % key)
    return df_to.merge(df_some_grades, how='outer', on='entityGuid')


df_final1 = df_ents_dense

for k in grade_columns.keys():
    df_final1 = merge_grade(df_grades, df_final1, k)
    
df_final1
```

```python
badge_columns = {
'overall' : 'best-places-to-live',
'healthiest' : 'healthiest-places',
'outdoor' : 'best-outdoors-cities',
'families' : 'best-places-for-families',
'safety' : 'safest-places',
'schools' : 'places-with-the-best-public-schools',
'diverse' : 'most-diverse-places',
'young_professionals' : 'best-places-for-young-professionals',
'retire' : 'best-places-to-retire',
'cost' : 'cost-of-living',
'housing' : 'best-places-to-buy-a-house'
}


def merge_badge(df_old, df_new, key):
    badgeType = badge_columns[key]
    df_some_badges = df_old.loc[df_old.badgeType == badgeType,:]
    df_some_badges = df_some_badges.drop_duplicates()

    df_some_badges.columns = df_some_badges.columns.str.replace('ordinal', 'rank_%s' % key)
    df_some_badges.columns = df_some_badges.columns.str.replace('total', 'rankTotal_%s' % key)
    df_some_badges = df_some_badges.drop(['badgeType'], 1)
    return df_new.merge(df_some_badges, how='outer',on='entityGuid')


df_final2 = df_final1

for k in badge_columns.keys():
    df_final2 = merge_badge(df_badges, df_final2, k)

df_final2 = df_final2.loc[df_final2['type_' ] != 'n']
df_final2
```

```python
#In order to deal with data easily and preventing the confusion, some columns are dropped.

rank_cols = df_final2.columns.str.startswith('rank')
df_niche_grades = df_final2.drop(df_final2.columns[rank_cols], 1)
```

```python
df_niche = pd.merge(df_niche_grades, df_facts, on='entityGuid')
df_niche.columns = df_niche.columns.str.replace('name', 'Town').str.replace ('tagline', 'Tagline')
df_niche = df_niche.drop('type_', 1)
df_niche 
```

```python
sns.set(style="whitegrid")

#Use semantically meaningful titles for the columns
col_names = df_niche.columns[6:17]


#Make the PairGrid
g = sns.PairGrid(df_niche.dropna().sort_values("Overall Niche Grade", ascending=False).head(50),
                 x_vars=col_names, y_vars=["Town"],
                 size=10, aspect=.25)

g.map(sns.stripplot, size=10, orient="h",
      palette="Reds_r", edgecolor="gray")


#Use the same x axis limits on all columns and add better labels
g.set(xlim=(0, 5), xlabel="Grade", ylabel="")



for ax, title in zip(g.axes.flat, col_names):

    if title != col_names[0]:
        ax.yaxis.set_visible(False)
       
    # Set a different title for each axes
    ax.set(title=title)

    # Make the grid horizontal instead of vertical
    ax.xaxis.grid(True)
    ax.yaxis.grid(False)

sns.despine(left=True, bottom=True)
```
